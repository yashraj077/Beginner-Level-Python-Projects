```python
# -*- coding: utf-8 -*-
"""
Created on Fri Jun  5 19:43:45 2020

@author: yash
"""


month_dict = {1:"January", 2:"February", 3:"March", 4:"April",
              5:"May", 6:"June",7:"July", 8:"August",
              9:"September", 10:"Octomber", 11:"November", 12:"December"}

month_days = {1:31, 2:29, 3:31, 4:30,
              5:31, 6:30, 7:31, 8:31,
              9:30, 10:31, 11:30, 12:31}

month_val = {1:0, 2:3, 3:3, 4:6,
             5:1, 6:4, 7:6, 8:2,
             9:5, 10:0, 11:3, 12:5}

def isLeap(year):
    if year%4==0:
        if year%100==0:
            if year%400==0:
                return 1
            else:
                return 0
        else:
            return 1
    else:
        return 0

def first_day(month, year, leap):
    date = 1
    if year>=1800 and year<=1899:
        century = 3
    elif year>=1900 and year<=1999:
        century = 1
    else:
        century = 0
    
    ans = date + month_val[month] + (year%100)-1 + ((year%100)-1)//4 + century
    
    ans = ans%7
    
    if month==1 or month==2:
        return ans+1
    elif leap==1:
        if ans==0:
            return 2
        elif ans==6:
            return 1
        else:
            return ans+2
    else:
        if ans==0:
            return 1
        else:
            return ans+1
    

def calendar(month, year, start, leap):
    count = start
    print("    S    M    T    W    T    F    S")
    
    print(" "*4, end="")
    print("     "*(count-1), end="")
    
    if month==2 and leap==1:    
        md = month_days[month]+1
    elif month==2 and leap==0:
        md = month_days[month]
    else:
        md = month_days[month]+1
        
    for i in range(1, md):
        if count == 7:
            count = 1
            print(i, end = "\n")
            if i>=9:
                print(" "*3, end = "")
            else:    
                print(" "*4, end = "")
        else:
            if i>=10:
                print(i, end="   ")
                count+=1
            elif i==9:
                print(i, end="   ")
                count+=1
            else:
                print(i, end="    ")
                count+=1

if __name__ == "__main__":
    month = int(input("Enter month (1-12) :"))
    year = int(input("Enter year (yyyy) :"))
    try:
        print("  ", month_dict[month], year)
    
        if month==2 and isLeap(year)==1:
            calendar(month, year, first_day(month, year, isLeap(year)), 1)
        else:
            calendar(month, year, first_day(month, year, isLeap(year)), 0)
    except:
        print("Enter month and year correctly")
```
